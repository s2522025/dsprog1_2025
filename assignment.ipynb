```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time
```

```python
# アクセスするURL
start_url = "https://www.musashino-u.ac.jp/"
```

```python
# ドメイン名
start_url = "https://www.musashino-u.ac.jp/"
domain = "musashino-u.ac.jp"
visited = set()
site_map = {}  # サイトマップを格納する辞書
MAX_PAGES = 50  # 最大アクセスページ数
SLEEP_TIME = 1  # アクセス間隔（秒）
```

```python
def crawl(url):
    # 既に訪問済み、または最大ページ数に達した場合は終了
    if url in visited or len(visited) >= MAX_PAGES:
        return

    # 訪問済みとしてマーク
    visited.add(url)
    print(f"{len(visited)}: {url}")

    try:
        time.sleep(SLEEP_TIME)
        response = requests.get(url, timeout=10)
        response.encoding = response.apparent_encoding
    except Exception as e:
        print(f"Error accessing {url}: {e}")
        return

    soup = BeautifulSoup(response.text, "html.parser")

    # サイトマップにタイトルを追加
    title_tag = soup.find("title")
    title_text = title_tag.get_text().strip() if title_tag else "No title"
    site_map[url] = title_text

    # 再帰的にリンクをたどる
    for link in soup.find_all("a", href=True):
        href = link["href"]
        full_url = urljoin(url, href)
        parsed = urlparse(full_url)

        # ドメインが一致し、httpまたはhttpsスキームの場合にクロール
        if domain in parsed.netloc and parsed.scheme in ["http", "https"]:
            crawl(full_url)
```

```python
# 実行
crawl(start_url)

# サイトマップを表示
for url, title in site_map.items():
    print(url, "→", title)
```
